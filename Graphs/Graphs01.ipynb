#### Making Dataset

- We have used numpy to produce a normally distributed dataset which shape is (1000,1) using np.random.randn.
- After producing we assinged the column name while loading the dataset as pandas dataframe.

```python
df = pd.DataFrame(np.random.randn(1000),columns=['data'])
```

```python
df
```

#### using plot() method

- plot() method by default creates line chart.

```python
df.plot()
```

```python
df.plot(figsize=(20,10))
```

#### Histogram

- Histogram plots the graph, keeping data points on x-axis and frequency on y-axis
- It is used to know the frequency of the range of data points.

```python
df.plot(kind='hist')
```

#### Box plot

- Used for identifying the outliers in the dataset

```python
df.plot(kind = 'box',figsize = (20,10))
```

#### Area plot

```python
df.plot(kind='area',figsize=(20,10))
```

```python
df.plot(figsize=(20,10),alpha =0.5)
```

#### Scatter plot

- Used to identify the co-relation of two variables

```python
df.plot.scatter(x = 'sepal_length', y = 'petal_length')
```

#### Scatter plot with three variables

```python
df.plot.scatter(x = 'sepal_length', y = 'petal_length',c = 'sepal_width')
```

#### Pie chart 

- Use only when you have 4 to 5 categories in the feature

```python
df.drop(columns='species').iloc[0].plot(kind= 'pie')
```
