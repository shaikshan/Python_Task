### Data types

#### String

String is a textual data in the python..

For textual data or string we can use `' '` or `" "`.

The difference is if we have to use `'` in the string then we should use

`" "`
For Example:

#### To print bobby's World

method

```python
message = "Bobby's World"
print(message)
```

Here `print()` is a function or statement used to output the result of the programme.

`"message"` is called as the `variable` where we store the value

value can any `data type`

### finding the length of the string

it can be found using the function `len()`

```python
message = "Roshan is a Associate Software Engineer"

len(message)
```

### String slicing operations

```python
message = "Roshan is just a name"

print(len(message))
```

```python
print(message[0])

print(message[0:6])
```

It should be noted that while printing a portion of the string like

` message[0:6]` the first instance `'0'` is inclusive and second instance is `"6"` is exclusive..

```python
print(message[:6])

print(message[7:])
```

You will get a error `"index out of range"` if you try to slice with the number which is greater than the lenght of string

```python
print(message[25])
```

### Replace a portion of text in the string:

This can be done using `'replace'` method

`For Example`:

```python
message = 'Hello World'

new_message = message.replace('World','Universe')

print(new_message)
```

### String concatenation:

`For Example:`

```python
greeting = "Hello"
Name = "David"
message = greeting + ' '+ Name

print(message)
```

#### Formatted String

```python
greeting = 'Hello'
Name = 'David'

message = '{} {}, Welcome!'.format(greeting,Name)

print(message)
```

#### Syntax for formatted string:

```python
string_name = '{} {}, Welcome!'
output = string_name.format(greeting,Name)
print(output)
```

### Alternate Way using f string

```python
greeting = "Hello"
Name = "David"
message = f'{greeting} {Name}, Welcome!'

print(message)
```

### You can also use functions in formatted strings like

`For Example:`

```python
# Print a Welcome sentence with name in upper case.

greeting = "Hello"

Name = "David"

message = f'{greeting} {Name.upper()}, Welcome!'

print(message)
```

#### Type Casting

```python
num1 = 100
num2 = 200

print("integer addition operation output: ", num1+num2)

num1_string = str(num1)
print(type(num1_string))

num2_string = str(num2)
print(type(num2_string))
```

In the above code we have converted the `integers` into `string` data type

Now we can also use the string operations like:

```python
print("String concatenation output: ", num1_string+num2_string)
```
